```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
 
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')
pd.set_option('max_columns',500)
pd.set_option('max_rows',500)
%matplotlib inline
print('modules have imported')
```

```python
# check for NaN values with proprtion
missing_values = df_train.isnull().sum() /len(df_train) *100
# lets filter cols by cols > 0 
missing_count = missing_values[missing_values > 0]
missing_count.sort_values()
```

```python
# lets look to our target variable 
sns.distplot(df_train['SalePrice'])
plt.show()
```

```python
# as we see its not normally distributed , its skewed 
# lets scale it 
target = np.log(df_train['SalePrice'])
```

```python
#check
sns.distplot(target)
```

```python
# finding the numericla features 
numeric_features = df_train.select_dtypes(include=np.number)

# finding the categorical features 
cate_features = df_train.select_dtypes(exclude=np.number)
```

```python
numeric_features.shape
```

```python
cate_features.shape # convert to numeric 
```

```python
# remove ID col 
#del df_train['Id']
numeric_features.drop(columns='Id', inplace = True)
```

```python
#check 
numeric_features.columns
```

```python
# correlation b/w numeric features
plt.figure(figsize=(10,5))
corr = numeric_features.corr()
sns.heatmap(corr)
plt.show()
```

```python
# EDA
# when was the property sold ?!!
df_train.groupby(['YrSold','MoSold'])['Id'].count().plot(kind= 'bar', figsize=(16,5))
plt.title('Sold time')
plt.show()
```

```python
# 6 and 7 months 
```

```python
# where are the property located ?

df_train.groupby('Neighborhood')['Id'].count().\
    sort_values(). \
    plot(kind='bar', figsize=(12,6))
plt.title('Property location')
plt.show()
```

```python
# NAmes - CollgCr
```

```python
# plot all numeric features   

# Unpivot a DataFrame from wide to long format, optionally leaving identifiers set.
# tuple, list, or ndarray, optional Column(s) to unpivot. If not specified, uses all columns that are not set as id_vars.

all_numeric = pd.melt(df_train , value_vars=sorted(numeric_features)) 
fig = sns.FacetGrid(all_numeric , col = 'variable', col_wrap = 4 ,  # Multi-plot grid for plotting conditional relationships.
                                                                    # “Wrap” the column variable at this width
                    sharex= False , sharey=False)     
fig= fig.map(sns.distplot, 'value')  # To draw a plot on every facet, pass a function and the name of one or more columns in the dataframe
```

```python
# positive skewed and bars for discrete values 
```

```python
# convert those cols from numeric to cate.
cols = ['MoSold', 'YrSold', 'MSSubClass']
for col in numeric_features[cols]:
    numeric_features[col].apply(lambda x :str(x))
```

```python
# check
numeric_features.MSSubClass.dtype
```

```python
# plot cate. features 
all_cate = pd.melt(df_train , value_vars = sorted(cate_features))

fig = sns.FacetGrid(all_cate , col='variable', col_wrap = 4 ,
                   sharex = False , sharey=False)
plt.xticks(rotation = 'vertical')
[plt.setp(ax.get_xticklabels(), rotation = 60) for ax in fig.axes.flat] #iterator
fig = fig.map(sns.countplot, 'value')
fig.fig.tight_layout()
plt.show()

# Get the xaxis' tick labels.
#set property") and getp() to set and get object properties,
#tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area
# axes.flat => https://stackoverflow.com/questions/46862861/what-does-axes-flat-in-matplotlib-do
```
