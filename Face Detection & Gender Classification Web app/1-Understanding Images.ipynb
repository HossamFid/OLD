## Image Resizing

```python
# technique used to structure ur data 
# there are 2 types of resize : shrnik and enlarge 

# shrink the original image with  (width=800, height= 800)
# into image with resizing shape  (width = 300 , height = 300)

# read image and convert it into grayscale  
img_re = cv2.resize(gray_cv, (300,300), cv2.INTER_AREA) # interpolation technique
```

```python
img_re.shape
```

```python
plt.imshow(img_re, cmap='gray')
```

```python
 # while shrinking you might lose the info. depends on the data u are working on
```

```python
# enlarge the original image with (width=405, height= 608)
# into image with resizing shape  (width = 1000 , height = 1000)

img_en = cv2.resize(gray_cv, (1100,900), cv2.INTER_CUBIC)
```

```python
img_en.shape
```

```python
plt.imshow(img_en, cmap='gray')
```

```python
# cv2.INTER_CUBIC   SLOW & GOOD(EFFICIENT)
# cv2.INTER_LINEAR  FAST & OKAY
```
