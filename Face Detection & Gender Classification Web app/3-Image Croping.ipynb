```python
# apply for all images 
def extract_images(path, num, gender):
    
    img = cv2.imread(path)
    #print(img)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #print(gray)
    harr = cv2.CascadeClassifier('Data/haarcascade_frontalface_default.xml')
    faces = haar.detectMultiScale(gray,1.5,5)
    #print(faces)
    
    for x,y,w,h in faces:
        croped = img[y:y+h,x:x+w]
        
        if gender == 'male':
            cv2.imwrite(f'./Data/data/male_crop/{gender}_{num}.jpg', croped)
        else:
            cv2.imwrite(f'./Data/data/female_crop/{gender}_{num}.jpg', croped)
        
        
                                 
```

```python
#test
extract_images(men_path[0],1,'male')
```

```python
for i, path in enumerate(men_path):
    try:
        extract_images(path,i,'male')
        print(f'image info : {i}/{len(men_path)} processed successfully')
    except:
        print(f'failed to proccess at image no. {i}' )
```

```python
for i, path in enumerate(female_path):
    try:
        extract_images(path,i,'female')
        print(f'image info : {i}/{len(men_path)} processed successfully')
    except:
        print(f'failed to proccess at image no. {i}' )
```
